```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
base = pd.read_csv('kc_house_data.csv')
```

```python
sqft_living = base['sqft_living'].values
price = base['price'].values
```

```python
r_sqft_living = sqft_living.reshape(-1, 1)
```

```python
lm = LinearRegression()
```

```python
lm.fit(r_sqft_living, price)
```

```python
lm.score(r_sqft_living, price)
```

```python
predict = lm.predict(r_sqft_living)
```

```python
sns.scatterplot(sqft_living, price)
plt.plot(sqft_living, predict, c='red')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_squared_log_error
```

```python
print(mean_absolute_error(price, predict))
```

```python
print(mean_squared_error(price, predict))
```

```python
print(mean_squared_log_error(price, predict))
```
