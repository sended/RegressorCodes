```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
base = pd.read_csv('kc_house_data.csv')
```

```python
base.head()
```

```python
print(np.corrcoef(base.sqft_living, base.price))
sns.scatterplot(base.sqft_living, base.price)
plt.show()
```

```python
print(np.corrcoef(base.sqft_lot, base.price))
sns.scatterplot(base.sqft_lot, base.price)
plt.show()
```

```python
print(np.corrcoef(base.sqft_above, base.price))
sns.scatterplot(base.sqft_above, base.price)
plt.show()
```

```python
print(np.corrcoef(base.sqft_basement, base.price))
sns.scatterplot(base.sqft_basement, base.price)
plt.show()
```

```python
print(np.corrcoef(base.sqft_lot15, base.price))
sns.scatterplot(base.sqft_lot15, base.price)
plt.show()
```

```python
print(np.corrcoef(base.floors, base.price))
sns.barplot(base.floors, base.price)
plt.show()
```

```python
print(np.corrcoef(base.bedrooms, base.price))
defaultSize = plt.rcParams['figure.figsize']
sns.barplot(base.bedrooms, base.price)
plt.show()
```

```python
print(np.corrcoef(base.bathrooms, base.price))
plt.rcParams['figure.figsize'] = (15, 5)
sns.barplot(base.bathrooms, base.price)
plt.show()
```

```python
print(np.corrcoef(base.grade, base.price))
plt.rcParams['figure.figsize'] = defaultSize
sns.barplot(base.grade, base.price)
plt.show()
```
