```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_squared_log_error
```

```python
base = pd.read_csv('kc_house_data.csv')
```

```python
X = base[['sqft_living', 'sqft_above']]
```

```python
Y = base.price
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter3D(xs=X.sqft_living, ys=X.sqft_above, zs=Y, zdir='z', label='ys=0, zdir=z')
plt.xlabel('sqft living')
plt.ylabel('sqft above')
plt.show()
```

```python
lm = LinearRegression()
lm.fit(X, Y)
predict = lm.predict(X)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter3D(xs=X.sqft_living, ys=X.sqft_above, zs=Y, zdir='z', label='ys=0, zdir=z')
ax.plot_surface(X=X.sqft_living, Y=X.sqft_above, Z=predict.reshape(-1, 1), color='red')
plt.xlabel('sqft living')
plt.ylabel('sqft above')
plt.show()
```

```python
print(mean_absolute_error(Y, predict))
```

```python
print(mean_squared_error(Y, predict))
```

```python
print(mean_squared_log_error(Y, predict))
```
